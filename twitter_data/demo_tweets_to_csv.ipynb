```python
from classes import tweet_api
import tweepy
```

```python
import pandas as pd
```

```python
auth = tweepy.OAuthHandler(tweet_api.API_KEY, tweet_api.API_SECRET)
auth.set_access_token(tweet_api.ACCESS_TOKEN, tweet_api.ACCESS_TOKEN_SECRET)

api = tweepy.API(auth)
```

```python
class Linstener(tweepy.Stream):
    
    tweets = []
    limit = 100 # 추출데이터 수
    
    def on_status(self, status):
        self.tweets.append(status) # 새로운 트윗시, tweets 리스트에 붙이기 
        # print(status.user.screen_name + ": " + status.text)
        
        # 추출 수량 'limit' 설정값 이상시, 연결해지
        if len(self.tweets) == self.limit:
            self.disconnect()
        
stream_tweet = Linstener(tweet_api.API_KEY, 
                         tweet_api.API_SECRET, 
                         tweet_api.ACCESS_TOKEN, 
                         tweet_api.ACCESS_TOKEN_SECRET)

```

```python
# Stream by keywords

keywords = ['병원감염']

stream_tweet.filter(track=keywords)

# Create DataFrame

columns = ['User', 'Tweet']
data = []

for tweet in stream_tweet.tweets:
#    if not tweet.truncated:     
    data.append([tweet.user.screen_name, tweet.text])
#        else:
#            data.append([tweet.user.screen_name, tweet.extended_tweet['full_text']])

df = pd.DataFrame(data, columns = columns)
# print(df.shape)
# df.head()
```

```python
# user tweets

keywords = '병원감염'
limit = 300

tweets = tweepy.Cursor(api.search_tweets, 
                       q=keywords,
                       count=100,
                       tweet_mode = 'extended').items(limit)

# Create DataFrame
columns = ['Time', 'User', 'Tweet']
data = []

for tweet in tweets:
    data.append([tweet.created_at, tweet.user.screen_name, tweet.full_text])

df=pd.DataFrame(data, columns=columns)
# print(df.shape)
# df.head()

df.to_csv('병원감염트윗.csv')
```
