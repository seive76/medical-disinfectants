```python
from urllib.request import urlopen 
from bs4 import BeautifulSoup
```

```python
html = urlopen("https://news.naver.com/")
bsObject = BeautifulSoup(html, "html.parser")
```

```python
result = {}

for link in bsObject.find_all('a'):
    result[link.text] = link.get('href')
```
